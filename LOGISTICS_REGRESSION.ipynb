**1. What is Simple Linear Regression (SLR)? Explain its purpose?**

  -  Simple Linear Regression (SLR) is a statistical method used to model the relationship between one independent variable (X) and one dependent variable (Y) by fitting a straight line through the data. SLR tries to find the best-fitting line (called the regression line) that predicts Y based on X.

     Purpose of Simple Linear Regression:


     1. Understand Relationships:


         Helps determine how one variable affects another.


         Example: How study hours affect exam scores.


      2.  Make Predictions:


           You can predict future values of Y using X.


          Example: Predict sales based on advertising spend.


     3.  Identify Trends:

         Shows whether the relationship is positive, negative, or neutral.

          Positive slope → both variables increase together

          Negative slope → one increases while the other decreases

     4.  Quantify Strength of Relationship:


         Measures how strongly X influences Y

**2.   What are the key assumptions of Simple Linear Regression?**

  -  Simple Linear Regression (SLR) works well only when certain assumptions are satisfied. These assumptions ensure that the model’s results (like predictions and coefficients) are valid and reliable.

     Key Assumptions of Simple Linear Regression :

  1. Linearity :

     -The relationship between X (independent variable) and Y (dependent variable) should be linear.

     -This means the data should roughly form a straight-line pattern.

  2. Independence of Errors :

      -The residuals (errors) should be independent of each other.

      -No correlation between consecutive errors.

  3. Homoscedasticity (Constant Variance):

       -The spread of residuals should be constant across all values of X.

      -Errors should not increase or decrease as X changes.

  4. Normality of Errors:

     -The residuals should be normally distributed.

     -This is important for hypothesis testing and confidence intervals.

  5. No Significant Outliers :

     -There should be no extreme values that heavily influence the regression line.

     -Outliers can distort slope and intercept.

  6. No Measurement Error in X (Independent Variable) :

     -The independent variable X is assumed to be measured accurately.

     -Errors should only be in Y, not X.

**3.   Write the mathematical equation for a simple linear regression model and
explain each term?**

  - The equation of a Simple Linear Regression (SLR) model iS :

      Y = β0 + β1(x) + ε

      Y (Dependent Variable) :

     -The output or response variable we want to predict.

     -Example: Marks, sales, salary, etc

     X (Independent Variable) :

     -The input or predictor variable used to explain Y.

     -Example: Study hourss, advertising cost, experience, etc

      β0(Intercept) :

      -The value of Y when x = 0

      -It represents the starting point of the regression line.

      β1(Slope or Coefficient) :

      -Shows how much Y changes when X increases by 1 unit.

      -It represents the rate of change.

      ε (Error Term / Residual) :

     -The difference between actual value and predicted value.

     -Captures factors not included in the model (random noise)

**5.  What is the method of least squares in linear regression?**

  -  The method of least squares is a technique used to find the best-fitting line in Simple Linear Regression.

     It chooses the line that minimizes the total error between the actual values and the predicted values.

     Instead of just minimizing errors, it minimizes the sum of the squares of the errors

     -Draw a line through data points

     -Measure how far each point is from the line

     -Square those distances

     -Add them up

     -Choose the line with the smallest total.

**6.  What is Logistic Regression? How does it differ from Linear Regression?**

  -  Logistic Regression is a statistical and machine learning method used for classification problems, where the output is categorical (usually 0 or 1).

     Instead of predicting a continuous value, it predicts the probability that an input belongs to a particular class.

     -Output is probability (0 to 1).

     -Uses an S-shaped (sigmoid) curve.

     -Suitable for binary classification.

     Difference Between Linear Regression and Logistic Regression:

     Linear Regression  :

     -Predict continuous values.
     
     -Output  →  Any real number
     
     -Use Case  → Price prediction, salary estimation

     -Error Minimization → Least Squares Method.

     Logistic Regression :

     -Predict categorical outcomes.

     -Output  →  Probability (0–1), then classified (0 or 1)

     -Use Case  → Spam detection, disease prediction

     Error Minimization → Maximum Likelihood Estimation

**7.  Name and briefly describe three common evaluation metrics for regression
models?**

  -  Common regression evaluation metrics include Mean Absolute Error (MAE) for average error magnitude, Root Mean Squared Error (RMSE) for penalizing large errors, and ( r square )Score for determining goodness-of-fit. These metrics help quantify how closely model predictions match actual, continuous data values.

  1.  Mean Absolute Error (MAE): Measures the average magnitude of errors in a set of predictions, without considering their direction (absolute differences between prediction and actual). It is highly interpretable because it is in the same units as the target variable.

  2.  Root Mean Squared Error (RMSE): Calculates the square root of the average squared differences between prediction and actual observation. By squaring the errors before averaging, RMSE heavily penalizes large errors or outliers compared to MAE.

  3.  R-Squared Score: Known as the coefficient of determination, this metric represents the proportion of variance in the dependent variable explained by the model. It ranges from minus infinity to 1 , where 1 indicates a perfect fit and 0 indicates the model performs no better than predicting the mean.

  **8.  What is the purpose of the R-squared metric in regression analysis?**

  -  R-squared or the coefficient of determination, measures the proportion of variance in the dependent variable explained by independent variables in a regression model. It indicates "goodness of fit"—how well the model fits data points (0 to 1 scale), allowing for comparison between models and aiding in assessing predictive accuracy.

     The purpose of the R-squared (R²) metric in regression analysis is to measure how well a model explains the variability in the dependent (target) variable.

     R² tells you how much of the variation in the actual data is captured by the model.

      -Evaluates goodness of fit.

     -Helps compare different regression models.

     -Indicates how well predictors explain the outcome.

**9.  Write Python code to fit a simple linear regression model using scikit-learn and print the slope and intercept.**

```python

import numpy as np
from sklearn.linear_model import LinearRegression

X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 4, 5, 4, 5])

model = LinearRegression()

model.fit(X, y)

slope = model.coef_[0]
intercept = model.intercept_

print("Slope (Coefficient):", slope)
print("Intercept:", intercept)
```
